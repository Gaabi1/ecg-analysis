## Filtragem

O sinal de ECG real vem cheio de ru√≠dos ‚Äî interfer√™ncias que n√£o t√™m origem el√©trica no cora√ß√£o.
A filtragem serve para limpar o sinal, removendo frequ√™ncias indesejadas e preservando apenas o conte√∫do fisiol√≥gico relevante.

Por que o ECG precisa ser filtrado?

O sinal do cora√ß√£o √© muito pequeno (na faixa de mV) e muito lento comparado a ru√≠dos.
Por isso, v√°rios tipos de interfer√™ncias aparecem:

Ent√£o, o filtro passa-banda (bandpass filter) √© usado para manter apenas as frequ√™ncias t√≠picas do ECG (entre 0,5 e 40 Hz).

#### O que o filtro passa-banda faz?

Ele deixa passar apenas o conte√∫do entre duas frequ√™ncias de corte:

S√≥ as frequ√™ncias entre 0.5 e 40 Hz passam.

#### Fun√ß√£o butter_bandpass_sos()

Essa fun√ß√£o cria o filtro Butterworth (um tipo cl√°ssico de filtro digital com resposta suave e sem ondula√ß√µes na banda √∫til).

- fs: taxa de amostragem (quantas amostras por segundo).
- nyquist: metade de fs ‚Äî √© o limite superior de frequ√™ncia represent√°vel sem gerar aliasing.
- low e high: frequ√™ncias normalizadas (0‚Äì1, em rela√ß√£o √† Nyquist).
- butter(): cria um filtro Butterworth passa-banda.
- output='sos': formato num√©rico mais est√°vel para implementar filtros (evita erros quando o filtro √© de ordem alta).

A fun√ß√£o sosfiltfilt() aplica o filtro duas vezes:

- uma vez para frente
- outra vez para tr√°s

Isso faz o atraso de fase ser cancelado, ou seja, os picos do ECG permanecem exatamente no mesmo tempo (sem ‚Äúatrasar‚Äù o batimento).
##### O que √© ‚Äúfase‚Äù num sinal

Todo sinal (como o ECG) tem amplitude e fase.

Amplitude ‚Üí o tamanho das ondas (quanto varia para cima e para baixo).

Fase ‚Üí o alinhamento temporal das partes da onda (em que momento ocorrem os picos).

Quando aplicamos um filtro digital, ele muda as amplitudes das frequ√™ncias ‚Äî e, muitas vezes, tamb√©m atrasa cada frequ√™ncia de forma diferente.

##### O problema do atraso de fase

Imagine que seu ECG tem um pico R (o ponto alto do batimento) exatamente em 1,00 segundo.
Se voc√™ aplicar um filtro com atraso de fase, esse pico pode aparecer deslocado, por exemplo, em 1,02 segundos.
